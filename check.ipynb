```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\monst\Desktop\Movie-Netflix-Recommendation\data\netflix_titles.csv")
```

```python
df.head(2)
```

```python
df['listed_in']=df['listed_in'].fillna("Unknown")
df.rename(columns={'listed_in':'genres'},inplace=True)
```

```python
df.fillna('',inplace=True)
```

```python
df['combined']=df['title']+' '+df['director']+' '+df['cast']+' '+df['genres']+' '+df['description']
df['combined']=df['combined'].str.lower()
df.head(4)
```

```python
df=df.drop(['show_id','type'],axis=1) #allready droped 
df.head(3)
```

```python
# droping the missing rows in 
df.dropna(subset=['title','description'],inplace=True)
```

```python
df['combined'].fillna("Data not found",inplace=True)
df.head(3)
```

```python
df.isnull().sum()
# no null values found 
# if there is a null value using .fillna it will be replaced by ''
```
